```python
# pandas is python library. it used tp work for data set 
import pandas as pd
```

```python
#ASign csv data to datset varible
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# Assign input value to independet variable
independent=dataset[["YearsExperience"]]
independent
```

```python
# Assign output alue  to dependent ariab;e
dependent=dataset[["Salary"]]
dependent
```

from sklearn.model_selection import train_test_split
train_test_split(independent,dependent,test_size=0.30,random_state=0)

```python
# data separate training data and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Find the model using Simple linear regression algorithms
rom sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
# assign w value or slop alue
weight = regressor.coef_
weight
```

```python
# assign inercept value or biasvalue to bais variable
bais=regressor.intercept_
bais
```

```python
# Predict test output value using test  input value
y_pred=regressor.predict(x_test)
```

```python
# Find the error value using r2 value
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
# save model  using pickle library
import pickle
filename="finalized_model_linear.sav"
```

```python
# n use pickle.dump() function to store the  data to the file. 
pickle.dump(regressor,open(filename,'wb'))
```
